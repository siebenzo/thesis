```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

df_tinyml = pd.read_excel('output_tinyml.xlsx', sheet_name='Blad1', nrows=101)

preprocessing_data = df_tinyml['preprocessing time']
face_detection_data = df_tinyml['face detection']
total_time_data = df_tinyml['total processing time']

#BOXPLOTS
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 6))

ax1.boxplot([preprocessing_data], showfliers=False, labels=['Preprocessing'])
ax1.set_ylabel('Time (ms)')
ax1.set_ylim(155, 158)

ax2.boxplot([face_detection_data], showfliers=False, labels=['Face detection'])
ax2.set_ylabel('Time (ms)')
ax2.set_ylim(2.99, 3.01)

ax3.boxplot([total_time_data], showfliers=False, labels=['Total time'])
ax3.set_ylabel('Time (ms)')
ax3.set_ylim(159, 161)

fig.patch.set_facecolor('white')
fig.suptitle('Boxplots tinyml')
plt.show()

# Print accuracy of the data
print(f'Preprocessing timing: {np.mean(preprocessing_data):.2f} ± {np.std(preprocessing_data):.2f} ms')
print(f'Face detection timing: {np.mean(face_detection_data):.2f} ± {np.std(face_detection_data):.2f} ms')
print(f'Total processing timing: {np.mean(total_time_data):.2f} ± {np.std(total_time_data):.2f} ms')

# SCATTERPLOTS
plt.scatter(range(len(preprocessing_data)), preprocessing_data, label='Preprocessing Time', color='green')
plt.scatter(range(len(face_detection_data)), face_detection_data, label='Face Detection Time', color='blue')

fig.patch.set_facecolor('white')
plt.title('Preprocessing vs Face detection Scatter Plot')
plt.xlabel('Sample')
plt.ylabel('Time (ms)')
plt.legend()

plt.show()

plt.scatter(range(len(total_time_data)), total_time_data, label='Total Time', color='red')

fig.patch.set_facecolor('white')
plt.title('Total processing Time Scatter Plot')
plt.xlabel('Sample')
plt.ylabel('Time (ms)')
plt.ylim(0,300)
plt.legend()

plt.show()
```
